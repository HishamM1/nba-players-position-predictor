```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

# Web Scraping player heights from ESPN player webpage

1. Read the links of NBA players' pages from links.xlsx file
2. Read the players' data on which we will append the heights from players_data.xlsx file

```python
links = pd.read_excel("links.xlsx")
players_data = pd.read_excel("players_data.xlsx")
links = links["stats_links"]
```

Perform web scarping using beautiful soup library 

```python
def get_heights(links):
    heights = []
    for index, link in enumerate(links):
        page = requests.get(link)
        soup = BeautifulSoup(page.content, "html.parser")
        # We select the html element the contains the player's height and weight => "7' 0", 280 lbs"
        # Split the string at "," => ["7' 0"", "280 lbs"]
        # Get the first element (the height) => 7' 0" and remove the " from its end => 7' 0 then split again at "' " => ["7", "0"]
        in_feet_inch = soup.select_one(".PlayerHeader__Bio_List li .fw-medium").text.split(",")[0].replace("\"","").split("' ")
        try:
            # convert the in_feet_inch list to a height in centimeter
            in_cm = int(in_feet_inch[0]) * 30.48 + int(in_feet_inch[1]) * 2.54
            # add it to the list of player heights 
            heights.append(in_cm)
        except:
            # if the above code fails make the height NaN
            heights.append(np.NAN)
    return heights

heights = get_heights(links)
# create a new column in the players' data table that contains the heights
players_data["height"] =  heights
```

```python
# Save the data with the heights
players_data.to_excel("players_data_with_heights.xlsx", index=False)
players_data
```
